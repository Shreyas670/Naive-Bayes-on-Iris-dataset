IMPORTING ESSENTIAL LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib.colors as colors
```

IMPORTING DATASET

```python
df=pd.read_csv('iris.csv')
```

VISUALISING HOW THE FLOWERS RELATE TO SEPAL  LENGTH AND WIDTH

```python
sns.FacetGrid(df, hue="Species", size=7) .map(plt.scatter,"SepalLengthCm", "SepalWidthCm", )  .add_legend()
plt.title('Scatter plot')
```

VISUALISING HOW THE FLOWERS RELATE TO PETAL LENGTH AND WIDTH

```python
sns.FacetGrid(df, hue="Species", size=7) .map(plt.scatter,"PetalLengthCm", "PetalWidthCm", )  .add_legend()
plt.title('Scatter plot')
```

```python
df=df.drop(['Id'],axis=1)
```

```python
df.head()
```

CATEGORISING DATASET INTO DEPENDENT INDEPENDENT AND TRAIN TEST 

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y=df.Species
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.75,random_state=100)
```

APPLYING GAUSSIAN NAIVE BAYES ALGO

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
```

PREDICTING TYPE OF IRIS AND PROBABLITIES OF A IRIS BEING ANY OF THE THREE

```python
Y_pred = model.predict(X_test)
```

```python
Y_prob = model.predict_proba(X_test)
Y_prob[-8:].round(2)
```

FIRST COL-PROBABILITIY THAT THE FLOWER IS SETOSA
SECOND COL-PROBABILITIY THAT THE FLOWER IS VERSICOLOR
THIRS COL-PROBABILITIY THAT THE FLOWER IS VIRGINIA

```python
from sklearn.metrics import accuracy_score
print(accuracy_score( Y_test, Y_pred))
print(np.mean(Y_test==Y_pred))
```

PLOTTING THE DECISION BOUNDARY BY CONSIDERING ONLY TWO FEATURES.SEPAL LENGTH AND WIDTH AS WE CAN VISUALISE ONLY 2D

```python
X_data = df.iloc[:,1:3]
y_labels = df.iloc[:,-1].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}).copy()

#Fit model
model_sk = GaussianNB(priors = None)
model_sk.fit(X_data,y_labels)


# Our 2-dimensional classifier will be over variables X and Y
N = 100
X = np.linspace(4, 8, N)
Y = np.linspace(1.5, 5, N)
X, Y = np.meshgrid(X, Y)

#fig = plt.figure(figsize = (10,10))
#ax = fig.gca()
color_list = ['Blues','Greens','Reds']
my_norm = colors.Normalize(vmin=-1.,vmax=1.)

g = sns.FacetGrid(df, hue="Species", size=10, palette = 'colorblind') .map(plt.scatter, "SepalLengthCm", "SepalWidthCm",)  .add_legend()
my_ax = g.ax


#Computing the predicted class function for each value on the grid
zz = np.array(  [model_sk.predict( [[xx,yy]])[0] for xx, yy in zip(np.ravel(X), np.ravel(Y)) ] )


#Reshaping the predicted class into the meshgrid shape
Z = zz.reshape(X.shape)


#Plot the filled and boundary contours
my_ax.contourf( X, Y, Z, 2, alpha = .1, colors = ('blue','green','red'))
my_ax.contour( X, Y, Z, 2, alpha = 1, colors = ('blue','green','red'))

# Addd axis and title
my_ax.set_xlabel('Sepal length')
my_ax.set_ylabel('Sepal width')
my_ax.set_title('Gaussian Naive Bayes decision boundaries')

plt.show()
```
